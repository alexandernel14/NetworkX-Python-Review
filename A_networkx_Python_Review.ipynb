```
# First lets import some useful packages to aid us in this NetworkX tutorial.
import sys
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```
# Initialize a basic empty graph (undirected graph)
Graph = nx.Graph()

# Add edges to the graph note that no nodes have been specified but it automatically creates nodes and edges for the graph
Graph.add_edge(0,1)
Graph.add_edge(1,2)

# Draw the graph and plot it
nx.draw(Graph)
plt.show()

# Print the basic information of the graph
print(nx.info(Graph))
```

```
# Create a new graph G.
G = nx.Graph()
G.add_edges_from([(1,2),(1,4),(1,6),(2,4),(2,3),(3,4),(3,7),(4,5),(5,6),(5,7),(6,7)])

# Draw the graph including its labels
nx.draw(G,with_labels= True)
plt.show()
```

```
# Get some basic information of this graph.
print(nx.info(G))
print(G.degree(1))

# Write a for loop to print the names and degrees of all the nodes in the graph G
for i in range(1,len(G.nodes)+1):
  print("Node " + str(i) + "'s degree is: ", G.degree(i))

# Define a general function to iterate over and print the names and degrees of all the nodes in any graph.
def NodeDegreeCalc(Graph):
  print("Degrees of nodes in the graph are:")
  for i in range(1,len(Graph.nodes)+1):
    print("Node " + str(i) + "'s degree is: ", Graph.degree(i))
```

```
# A complete useful information calculator for a graph
def NodeDegreeCalcAdvanced(Graph):
  # Draw the Graph
  nx.draw(Graph,with_labels= True)
  plt.show()

  print("Degrees of nodes in the graph are:")
  ListOfNodes = [None]*(len(Graph.nodes))

  for i in range(1,len(Graph.nodes)+1):
    print("Node " + str(i) + "'s degree is: ", Graph.degree(i))
    ListOfNodes[i-1] = Graph.degree(i)
  #print(ListOfNodes)

  Average_Degrees = sum(ListOfNodes)/len(Graph.nodes)
  print("The average degrees in the entire graph is: " + str(Average_Degrees))

  L = nx.normalized_laplacian_matrix(Graph)
  e = np.round(np.linalg.eigvals(L.A),1)
  print(e)
  print("Largest eigenvalue:", max(e))
  print("Smallest eigenvalue:", min(e))
  # Get the number of eigenvalues that are equal to 0 this will give the number of distinct graphs.
  NumberGraphs = len(e[e == 0])
  print("The number of eigenvalues with value of 0 is: " +str(NumberGraphs) + ". Thus, there are " +  str(NumberGraphs) + " distinct graphs.")
  plt.hist(e)
  plt.title("Histogram of Eigenvalues")
  plt.xlabel("Eigenvalue")
  plt.ylabel("Count")
  plt.xlim(0, max(e))
  plt.show()

```

```
# Question 2 (a)

def PageRankCalc(Graph,alpha,iter):
  # Calculate the Adjacency Matrix (make use of nx.adjacency_matrix(Graph))
  AdjacencyMat = nx.adjacency_matrix(Graph,nodelist= range(1,1+len(Graph)),weight="weight")

  # Create a list containing all the values of the page rank vector r
  r = [None]*(iter+1)
  r[0] = (1/len(Graph.nodes()))*np.ones((len(Graph.nodes()),1))

  # Create a list to collect all the outdegrees of all the nodes in the DGQ1 graph.
  ListOfOutDegrees = [None]*(len(Graph.nodes))

  for i in range(1,len(Graph.nodes)+1):
    ListOfOutDegrees[i-1] = Graph.out_degree(i)

  # Initialize a n by n matrix of zeros to be used to construct the out degree matrix.
  D = np.zeros((len(Graph.nodes),len(Graph.nodes)))
  for i in range(0,len(Graph.nodes)):
    D[i,i] = ListOfOutDegrees[i]

  # Calculate the inverse of the D matrix.
  D_Inv = inv(D)

  # Calculate the Normalized Adjacency Matrix.
  N = np.matmul(inv(D),AdjacencyMat.todense())

  # Used a for loop to carry out the iterations
  for i in range(1,iter+1):
    r[i] = alpha*np.matmul(np.transpose(N),r[i-1]) + (1-alpha)*(1/len(Graph.nodes))*np.ones((len(Graph.nodes),1))
  print(" The Page Rank Vector after iterating " +str(iter) + " times is: \n",r[iter],"\n")

  # Sort and print the names of the nodes and their corresponding Page Ranks for the top 3 nodes.
  # Get the names of all the nodes in the graph
  ListOfNodes = list(Graph.nodes)

  #Create a DataFrame to contain all the nodes and their respective page ranks.
  df_Page_Rank_Scores = pd.DataFrame(ListOfNodes,columns= ["Node_Id"])
  df_Page_Rank_Scores["Page_Rank_Scores"] = r[iter]

  # Sort the DataFrame in descending order to easily access the top three elements.
  df_Page_Rank_Scores.sort_values(by=['Page_Rank_Scores'],ascending = False,inplace = True)

  # After sorting print the top three nodes and their page rank scores
  print("The top three nodes by Page Rank are: \n",df_Page_Rank_Scores.iloc[0:3,:].to_string(index = False))
```

```
# Call the function on the appropriate graph with alpha = 0.8 and iter = 20.
PageRankCalc(DGQ2,0.8,20)
```
